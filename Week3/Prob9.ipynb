```python
import os
from src import mylib

keyPath = os.path.join(os.getcwd(), 'src', 'key.properties')
key = mylib.getKey(keyPath)
```

```python
SERVICE = 'AccidentDeath'
OPERATION_NAME = 'getRestTrafficAccidentDeath'
HEAD = os.path.join(SERVICE, OPERATION_NAME)
print(HEAD)
```

```python
import urllib

_d = dict()
_d['searchYear'] = '2020'
_d['siDo'] = '1100' # 서울 특별시
_d['guGun'] = '1116'
_d['numOfRows'] = '20'
_d['pageNo'] = '1'

QUERY = urllib.parse.urlencode(_d)
print(QUERY)
```

```python
params = HEAD + '?serviceKey=' + key['gokr'] + '&' + QUERY
print(params)
```

```python
endpoint = 'http://apis.data.go.kr/B552061/'
url = urllib.parse.urljoin(endpoint, params)
print(url)
```

```python
import requests

data = requests.get(url).text
print(data)
```
