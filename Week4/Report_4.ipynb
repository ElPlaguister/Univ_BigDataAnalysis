```python
import os
import pyspark
```

```python
myConf = pyspark.SparkConf()
spark = pyspark.sql.SparkSession\
.builder\
.master('local')\
.appName('myApp')\
.config(conf=myConf)\
.getOrCreate()
```

```python
bd_rdd = spark.sparkContext.textFile(os.path.join('data', 'ds_bigdata_wiki.txt'), use_unicode=True)
bd_rdd.collect()
```

```python
bd_rdd_base = bd_rdd.flatMap(lambda line: line.split())
bd_rdd_base.collect()
```

```python
bd_list = bd_rdd_base.map(lambda word: (word, 1))\
.groupByKey()\
.mapValues(sum)\
.sortByKey(True)\
.collect()
bd_list
```

```python
bd_list_sorted = sorted(bd_list, key = lambda element: -element[1])
bd_list_sorted
```

```python
for element in bd_list_sorted[:15]:
    _key = element[0]
    _value = element[1]
    print (f"{_key}\t{_value}")
```
